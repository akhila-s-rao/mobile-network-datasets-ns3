# Create a REM map for a given rem file

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
#===================================
# Data source
#=================================== 
data_dir = '../../data_volume/'+'rem_log_downlink_rangeLoss/'
rem_file='rem.out'
print(data_dir+'run1/'+rem_file)
```

```python
df = pd.read_csv(data_dir+'run1/'+rem_file, sep='\t', header=None, names=['x', 'y', 'z', 'sinr_linear'])
print(df.shape)
print(df.head)
#print(df['sinr_linear'].describe())
print(df[['x', 'y']])
```

```python
fig = plt.figure(figsize=(15,15))

data = 10*np.log10(df['sinr_linear'].to_numpy()).reshape(100, 100)

plt.imshow(data,interpolation='none',origin='lower')
plt.colorbar()

#plt.pcolor(df[['x', 'y']], 10*log10(df['sinr_linear']))
#plt.show()
#ax = sns.heatmap(corr_data, cmap=sns.color_palette("coolwarm", as_cmap=True), linewidth=.5, fmt=".1f", annot=True)
#ax.set_aspect("equal")
```
